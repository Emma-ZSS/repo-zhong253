```python
# Task 1 - Worksheet 2
# Purpose: What state has the fewest number of counties in it?
# Output: A single string value indict the state
```

```python
import pandas as pd
from pandas import DataFrame, Series
```

```python
census = pd.read_csv('census.csv', encoding = 'latin1') # keep index column
census.head()
#census.keys()
```

```python
def answer1():
    minv = census['COUNTY'].min()
    row = census.loc[census['COUNTY']==minv].index[0] # .index[0] return subset string(row index)
    return census.loc[row,'CTYNAME'] # Select rows and columns
answer1()
```

```python
# Task 2 - Worksheet 2
# Purpose: Which state has the highest average growth in population due to births over the 5 year span 
#    from 2010 to 2015?
# Output: The state and the average growth due to births over the 5 year span.
```

```python
def answer2():
    # New dataframe: State (full state)
    state = census[census['SUMLEV']==40][['STNAME','BIRTHS2010', 'BIRTHS2011', 'BIRTHS2012', 'BIRTHS2013', 'BIRTHS2014','BIRTHS2015']]
    # Birth rate (5 year span)
    g_2011=(state['BIRTHS2011']-state['BIRTHS2010'])/state['BIRTHS2010']
    g_2012=(state['BIRTHS2012']-state['BIRTHS2011'])/state['BIRTHS2011']
    g_2013=(state['BIRTHS2013']-state['BIRTHS2012'])/state['BIRTHS2012']
    g_2014=(state['BIRTHS2014']-state['BIRTHS2013'])/state['BIRTHS2013']
    g_2015=(state['BIRTHS2015']-state['BIRTHS2014'])/state['BIRTHS2014']
    # Average birth rate
    state['AverageGrowth']=(g_2011+g_2012+g_2013+g_2014+g_2015)/5
    # Highest average birth rate
    highest = state['AverageGrowth'].max()
    # Get index & Output
    ind = state.loc[state['AverageGrowth']==highest].index[0]
    return state.loc[ind,['STNAME','AverageGrowth']]
answer2()
```
